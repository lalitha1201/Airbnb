## Applying Beautiful Soup

Webscrapping is a very important way to extract new dataset for better insight.
Basically, Beautiful Soup is a Python library for pulling data out of HTML and XML files.

Next, we can create a request session abd assign the result of a request of that page to the variable page with the request.get() method. We pass the page’s URL to that method. 

In this case, we are parsing information from zillow, specialized in Seattle City.

```python
from bs4 import BeautifulSoup
import requests

requirement = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.8',
    'upgrade-insecure-requests': '2',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}

with requests.Session() as s:
    url= 'https://www.zillow.com/homes/for_sale/seattle_rb/'
    r = s.get(url, headers=requirement)
```

In this section, we start our scrapping. We scrape for the features we want such as:,price, information and address.

Information which includes: bedrooms. bathrooms and how many square feet is the house.

```python
content_scrape = BeautifulSoup(r.content, 'lxml')
price_scrape = content_scrape.find('span', {'class': 'zsg-photo-card-price'}).text
infomation_scrape =content_scrape.find('span', {'class': 'zsg-photo-card-info'}).text
address_scrape = content_scrape.find('span', {'class': 'zsg-photo-card-address'}).text
```
