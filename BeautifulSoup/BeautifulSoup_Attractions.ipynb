```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains
driver = webdriver.Chrome()
actions = ActionChains(driver)

driver.get('https://www.visitseattle.org/things-to-do/sightseeing/top-25-attractions/')
```

```python
pghtml = driver.page_source
```

```python
#creates a Beautiful Soup scraper
from bs4 import BeautifulSoup
import requests

soup = BeautifulSoup(pghtml, "html.parser")
```

```python
soup
```

```python
block_class = soup.findAll('h3')

block_class
```

```python
attractions = []
for x in block_class:
    attraction = x.string
    attractions.append(attraction)
attractions
```

```python
attractions = attractions[:-1]
```

```python
coordinates = []
```

```python
# coordinates = coordinates.append(attractions)
# coordinates
```

```python
# print(coordinates)
```

```python
driver.get('http://www.latlong.net/')
```

```python
for attraction in attractions:
    time.sleep(3)
    search = driver.find_element_by_id('place')
    search.clear()
    time.sleep(3)
    search.send_keys(attraction)
    time.sleep(3)
    print()
    time.sleep(1)
    actions.send_keys(Keys.ARROW_DOWN).perform()
    time.sleep(3)
    
    print('sending enter')
    time.sleep(1)
    
    actions.send_keys(Keys.ENTER).perform()

    lat = driver.find_element_by_id('latlngspan')

    lat_long = lat.text.strip('() ').split(',')
    lat_long_clean = [float(n) for n in lat_long]
    lat_long_clean.append(attraction)
    print(lat_long_clean) 
    coordinates.append(lat_long_clean)
```
